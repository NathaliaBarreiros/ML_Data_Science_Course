```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
Temp_I = [0.72, 0.61, 0.65, 0.68, 0.75, 0.90, 1.02, 0.93, 0.85, 0.99, 1.02]
```

### Line plot

```python
# plt.plot(x,y)
plt.plot(Year, Temp_I)
plt.xlabel("Year")
plt.ylabel("Temperature Index")
plt.title("Global Warming", {'fontsize':12,'horizontalalignment':'center'})

plt.show()
```

#### Example 2

```python
Month = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Customer1 = [12,13,9,8,7,8,8,7,6,5,8,10]
Customer2 = [14,16,11,7,6,6,7,6,5,8,9,12]
```

```python
plt.plot(Month, Customer1, color='olive', label='Customer 1', marker='o')
plt.plot(Month, Customer2, color='orange', label='Customer 2', marker='>')
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Buildings Consumption")
plt.legend(loc='upper right')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(Month, Customer1, color='olive', label='Customer 1', marker='o')
plt.xlabel("Month")
plt.ylabel("Electricity Consumption")
plt.title("Buildings Consumption of Customer 1")
# plt.show()

plt.subplot(1,2,2)
plt.plot(Month, Customer2, color='orange', label='Customer 2', marker='>')
plt.xlabel("Month")
# plt.ylabel("Electricity Consumption")
plt.title("Buildings Consumption of Customer 2")
plt.show()
```

### Scatter plot

```python
plt.scatter(Month, Customer1, color='purple', label='Customer 1')
plt.scatter(Month, Customer2, color='green', label='Customer 2')
plt.ylabel("Electricity Consumption")
plt.title('Scatter plot of Building Consumption')
plt.grid()
plt.legend(loc='best')
plt.show()
```

### Histograms

Show us the distribution of the data
- It gives in x -axis the probability or # of ocurrence in each data

```python
plt.hist(Customer1, bins=10, color='orange')
plt.ylabel('Electricity Consumption')
plt.title('Histogram')
plt.show()
```

Interpretation: This graph gives the distribution of the data in all year
- The consumption between 7.5 & 8 is more than the consumption in 5 & 6
- The msot consumption is between 7.5 & 8
- We can find if this is a normal or uniform distribution using this graph

### Bar plot

In x-axis, it has the Month array
- This graph is like the scatter plot

```python
plt.bar(Month, Customer1, color='orange', width=0.8)
plt.show()
```

A good bar chart is when you can separate each column from each other

```python
plt.bar(Month, Customer1, color='purple', label='Customer 1')
plt.bar(Month, Customer2, color='green', label='Customer 2')
plt.ylabel("Electricity Consumption")
plt.title('Bar plot of Building Consumption')
plt.grid()
plt.legend(loc='best')
plt.show()

```

```python
bar_width = 0.2
Month_b = np.arange(12)
Month_b
Customer3 = [15,10,12,5,4,4,8,10,15,8,12,10]


plt.bar(Month_b, Customer1, bar_width, color='purple', label='Customer 1')
plt.bar(Month_b+bar_width, Customer2, bar_width, color='green', label='Customer 2')
plt.bar(Month_b+2*bar_width, Customer3, bar_width, color='orange', label='Customer 3')
plt.legend(loc='best')
plt.xlabel('Month')
plt.ylabel('Electricity Consumption')
plt.title('Bar Chart')

# Change the labels of x-axis
plt.xticks(Month_b + (bar_width)/12, ('Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))

plt.show()
```

### Box plot

```python
# vert = True -> vertical boxplot
# notch = True -> not simple rectangular form
# plt.boxplot(Customer1, notch=True, vert=False)
# orange line -> median
# final lines -> caps
plt.boxplot(Customer1, notch=False, vert=True)


plt.show()
```
